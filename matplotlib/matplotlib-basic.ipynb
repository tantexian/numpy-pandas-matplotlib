模拟一个随机散点图


````python
matrix_size = 300
x = rand(matrix_size)
y = rand(matrix_size)
size = rand(matrix_size)*50
color = rand(matrix_size)
# x,y,size.color为100的列向量
scatter(x,y,size,color)

# 显示颜色条
colorbar()
````

## 一次绘制多张图

````python
# 使用figure()命令产生新的图像：
x = linspace(0,3.14*2,100)
print(x)

plt.plot(x,sin(x))
figure()
plt.plot(x,2*sin(x))
figure()
plt.plot(x,sin(2*x))
````

````python
# 同一个图，绘制多个图形
# 使用 subplot 在一幅图中画多幅子图：
# subplot(row, column, index) row表示行，col表示列，index表示排在第几个区域绘图

subplot(2, 2, 1)
plt.plot(x,sin(x))
subplot(2, 2, 2)
plt.plot(x,2*sin(x))
subplot(2, 2, 4)
plt.plot(x,sin(2*x))
````

````python
# 绘图加入标签
plt.plot(x,sin(x),label="sin(x)")
plt.plot(x,2*sin(x),label="2*sin(x)")

# legend()加上标签图例(刻印文字)
legend()
````

````python
# 后加也可以
plt.plot(x,sin(x),label="sin(x)")
plt.plot(x,2*sin(x),label="2*sin(x)")
legend(['sin', 'cos'])
````
